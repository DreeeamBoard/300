```python
def get_balance(ticker):
    # 잔고 조회
    balances = upbit.get_balances()
    for b in balances:
        if b['currency'] == ticker:
            if b['balance'] is not None:
                return float(b['balance'])
            else:
                return 0
    return 0

get_balance('KRW')
```

```python
# 16강
# 변동성 돌파 전략 (THE Volatility break-out strategy)
# 레인지 : 전일 고가 - 전일 저가 (하루 안에 움직인 가격의 최대 폭)
# 매수 기준 : 오늘 시가 기준으로 가격이 '레인지 * k'이상 상승하면 (돌파하면) 해당 가격에 매수
# k는 0.5 ~ 1
# 매도 기준 : 그날 종가에 판다

# 백테스팅
df = pyupbit.get_ohlcv('KRW-BTC', 'day', 200)
print(df.head())
df.to_excel("btc.xlsx")

# 레인지를 계산하고
# 목표가 = 시가 + 0.5 * 레인지
# 매수 : 고가(high)가 목표가를 넘으면 1, 아니면 0 
```

```python
# 17강

# 수익률 계산
# 매수한 경우 : 매도가 / 매수가 = 종가 / 목표가
# 매수하지 않은 경우 : 수익률 1

# 총 수익률 = 수익률 모두 곱하기

# 2021-12-11 시가 59749000
# 2022-06-28 시가 27033000

# 단순히 보유했을 떄는 수익률이 0.452442719
# 변동성 돌파 전략 수익률은 0.926614515
```

```python
# 변동성 돌파 전략 구현해보기

df = pyupbit.get_ohlcv('KRW-BTC', 'day')
print(df.tail())

today = df.iloc[-1]
yesterday = df.iloc[-2]

yesterday_range = yesterday['high'] - yesterday['low'] # range 구하기
target = today['open'] + yesterday_range * 0.5 # 목표가 구하기
print(target)
```

```python
# 함수로 만들기

def cal_target(ticker):
    df = pyupbit.get_ohlcv(ticker, 'day')
    yesterday = df.iloc[-2]
    today = df.iloc[-1]
    yesterday_range = yesterday['high'] - yesterday['low']
    target = today['open'] + yesterday_range * 0.5
    return target
    
target = cal_target('KRW-XRP')
print(target)
```

```python
# 매일 아침 9시에 목표 갱신이 필요하다

target = cal_target('KRW-BTC')

while True :
    now = datetime.datetime.now()
    
    # 9시 0분 20초 ~ 30초 사이에 목표가 갱신
    if now.hour == 9 and now.minute == 0 and 20 <= now.second <= 30 :
        target = cal_target("KRW-BTC")
        time.sleep(10) # 계속 갱신하는 것 방지
    
    price = pyupbit.get_current_price('KRW-BTC')
    print(now, price)
    time.sleep(1)
```

```python
# 19강
# 변동성 돌파 전략 구현 - 2

# 프로그램을 처음 시작한 날은 매수되지 않도록 처리하기.
# 현재가가 목표가보다 한참 위에 있을 때 매수가 될 수 있기 때문이다

target = cal_target('KRW-BTC')
op_mode = False # 내일 아침 9시에 true가 되게 설정해주자

print(target)

while True :
    now = datetime.datetime.now()
    
    if now.hour == 9 and now.minute == 0 and 20 <= now.second <= 30 :
        target = cal_target('KRW-BTC')
        
        op_mode = True
        time.sleep(10)
        
    price = pyupbit.get_current_price('KRW-BTC')
    print(now, price)
    time.sleep(1)
```

```python
# 매수 기능 추가
# hold

target = cal_target('KRW-BTC')
op_mode = False # 내일 아침 9시에 true가 되게 설정해주자
hold = False

while True :
    now = datetime.datetime.now()
    
    if now.hour == 9 and now.minute == 0 and 20 <= now.second <= 30 : # 9시 00분 20초쯤이 되면 첫 거래가 발생하니까
        target = cal_target("KRW-BTC")
        op_mode = True
        
    price = pyupbit.get_current_price('KRW-BTC')
    
    # 매초마다 조건을 확인한 후 매수 시도
    if op_mode is True and hold is False and price >= target :
        # 잔고의 10%로만 매수
        krw_balance = upbit.get_balance("KRW")
        upbit.buy_market_order("KRW-BTC", krw_balance * 0.1)
        hold = True
        
    print(now, price)
    time.sleep(1)
```

```python
# 매도 기능 추가

while True :
    
    now = datetime.datetime.now()
    
    if now.hour == 8 and now.minute == 59 and 50 <= now.second <= 59 : # 8시 59분 50초 부근에
        if op_mode is True and hold is True :
            btc_balance = upbit.get_balance("KRW-BTC")
            upbit.sell_market_order("KRW-BTC", btc_balance)
            hold = False # 보유상태를 false로. 팔았으니까
            
        # 새로운 거래일에서 목표가 갱신될 때 까지 거래가 되지 않도록
        op_mode = False
        time.sleep(10)
```

```python
# 상태 출력 추가

# 매수 기능 추가
# hold

target = cal_target('KRW-BTC')
op_mode = False # 내일 아침 9시에 true가 되게 설정해주자
hold = False

while True :
    now = datetime.datetime.now()
    
    if now.hour == 9 and now.minute == 0 and 20 <= now.second <= 30 :
        target = cal_target("KRW-BTC")
        op_mode = True
        
    price = pyupbit.get_current_price('KRW-BTC')
    
    # 매초마다 조건을 확인한 후 매수 시도
    if op_mode is True and hold is False and price >= target :
        # 잔고의 10%로만 매수
        krw_balance = upbit.get_balance("KRW")
        upbit.buy_market_order("KRW-BTC", krw_balance * 0.1)
        hold = True
        
    # 상태 출력
    print(f"현재시간 : {now} 목표가 : {target} 현재가 : {price} 보유상태 : {hold} 동작상태 : {op_mode}")
    time.sleep(1)
    
# 이까지 406줄
```
